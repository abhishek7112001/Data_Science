## 🔹 Introduction to Generative AI & LLMs

* LLMs learn statistical language patterns from **massive text datasets**
* They can: summarise, code, translate, classify, retrieve info
* Output = Natural language **completion**
* Input = **Prompt** inside a **context window**
* Inference = generating new tokens based on previous tokens
* Larger models = improved generalization and **emergent capabilities**

💡 Bigger ≠ Always Better → depends on use case

---

## 🔹 Transformer Architecture – Why It Matters

RNN limitations: sequential processing, short memory, slow, struggle with context/ambiguity
Transformers solve this with:

### Key Components

* **Tokenization** → convert text → numbers
* **Embedding layer** → vector representation of meaning
* **Positional encoding** → keeps word order
* **Self-attention** → each token attends to all others
* **Multi-head attention** → different heads learn different relationships
* **Feed-forward network** → produce scores for next token
* **Softmax** → probabilities of next word

📌 Supports **parallelization**, scalable on GPUs → breakthrough!

---

### 🧠 Self-Attention Intuition

* Learns **which words influence each other**
* Enables understanding of **long-distance dependencies**
* Improves translation, reasoning, summarization

📊 Example: *“The teacher taught the students with the book”*
Attention maps help model decide **who has the book** ✔

---

## 🔹 Transformer Model Types

| Model Type      | Architecture   | Strengths                      | Examples                 |
| --------------- | -------------- | ------------------------------ | ------------------------ |
| Encoder-only    | Autoencoding   | Classification, sentiment, NER | BERT, RoBERTa            |
| Decoder-only    | Autoregressive | Text generation                | GPT family, BLOOM, LLaMA |
| Encoder-Decoder | Seq-to-seq     | Translation, summarization     | T5, BART                 |

---

## 🔹 Prompt Engineering

Method to shape model behavior using prompts

* **Zero-shot** → only instruction
* **One-shot** → add 1 example
* **Few-shot** → multiple examples
* **In-context learning** improves smaller model performance

🔑 Prompt = Instruction + data + output style hint

---

### 🧪 Sampling & Generation Controls

| Parameter           | Effect                                             |
| ------------------- | -------------------------------------------------- |
| **max_new_tokens**  | limits length of output                            |
| **temperature**     | randomness (high → creative, low → factual)        |
| **top-k**           | restrict choice to k highest-probability words     |
| **top-p (nucleus)** | restrict to words contributing up to p probability |

📎 Greedy decoding = always pick highest probability → repetitive!

## 🔹 Why Fine-Tune?

Prompt engineering alone **might not deliver accuracy** needed for:

* Domain-specific applications
* Complex classification
* Consistent structured outputs
* Smaller models with weak zero-shot skills

Fine-tuning = additional training on **task-specific data** 🧠
→ Model learns required behavior **in weights**, not just from the prompt context

---

## 🔹 How Models Are Initially Trained (Pre-Training)

> Builds **general world knowledge + language understanding**

Data: **Huge corpora** scraped + curated
Training: **Self-supervised** (no human labels)
Outcome: Foundational linguistic intelligence

But…
⚠️ Can't yet perform custom business tasks optimally
→ That’s where fine-tuning comes in

---

## 🔹 Fine-Tuning Approaches

| Approach                                   | Description                                        | Pros                             | Cons                                    |
| ------------------------------------------ | -------------------------------------------------- | -------------------------------- | --------------------------------------- |
| **Full fine-tuning**                       | Update *all* parameters                            | Maximum performance              | Very expensive                          |
| **PEFT (Parameter Efficient Fine-Tuning)** | Update *small subset* (LoRA adapters etc.)         | Fast, low cost, memory-efficient | Slightly lower max performance          |
| **Instruction fine-tuning**                | Fine-tune model to follow human-style instructions | Better chat-like behavior        | Needs high-quality labeled instructions |

💬 Popular PEFT method → **LoRA (Low-Rank Adaptation)**
Only trains *tiny* matrices inserted into attention layers → 1000x cheaper!

---

## 🔹 Model Architectures & Training Objectives

Which model to choose depends on:

* Input/output format
* Task structure
* Data available

| Architecture        | Training Objective                 | Typical Uses               |
| ------------------- | ---------------------------------- | -------------------------- |
| **Encoder-only**    | Masked language modeling           | Detection, classification  |
| **Decoder-only**    | Next-token prediction              | Generative chat + code     |
| **Encoder-Decoder** | Seq-to-seq (span corruption, etc.) | Summarization, translation |

Example from lab uses **T5 (seq-to-seq)** → ideal for summarization!

---

## 🔹 Data Requirements for Fine-Tuning

* Dataset format must match **task goal**
* High-quality labeled samples → huge uplift
* Alignment between **tokenizer** used in training & inference required
* Consider context window limits during data creation

📌 Even **hundreds to thousands** of good samples can beat billion-token pre-training for a specific task!

---

## 🔹 Choosing a Model for Your Use Case

Key considerations:

* Do you need broad knowledge? → large model
* Only need a focused workflow? → smaller fine-tuned model
* Real-time app constraints? → latency and memory matter
* Evaluate performance before training further

💡 Many times **small adapted models > giant generic models** for enterprise apps
→ Cheaper • Faster • Private

# 📘 Week 3 – Alignment & Reinforcement Learning from Human Feedback (RLHF)

---

## 🎯 Week 3 Learning Goals

* Understand **why** LLMs misbehave
* Learn **alignment techniques** beyond fine-tuning
* Reinforcement Learning from Human Feedback (RLHF)
* How to build a **reward model** for safer outputs
* Evaluate model behavior based on **HHH principles**

---

## 🔥 Why Alignment Is Needed

Even after fine-tuning…

LLMs **can still**:
❌ Use harmful / toxic language
❌ Provide confident but wrong answers
❌ Give dangerous instructions
❌ Generate hallucinations
❌ Fail ethical expectations

Cause:
Trained on **Internet-scale data** where harmful content exists

→ We must **guide behavior** closer to **human values**

---

## 🧭 HHH — Alignment Principles

| Principle    | Goal                            |
| ------------ | ------------------------------- |
| **Helpful**  | Useful + follows instructions   |
| **Honest**   | Truthful + avoids hallucination |
| **Harmless** | No offense, no risky info       |

These become **reward signals** in RLHF training.

---

## 🔹 What RLHF Does

Moves the model toward **human-preferred** responses
by learning from **human ratings** of outputs

3 Core Stages:

```mermaid
flowchart TD
A["""Supervised Fine-tuning"""] --> B["""Reward Model Training"""]
B --> C["""RL Optimization
(PPO etc.)"""]
```


### 1️⃣ Supervised Fine-Tuning (SFT)

* Humans write **ideal responses**
* Model trained to imitate them → “Teacher phase”

### 2️⃣ Reward Model (RM)

* Humans rank multiple model outputs
* Learn a scoring function that predicts “goodness”

### 3️⃣ RL Optimization (PPO)

* The model generates → RM scores it → gradients update policy
* Iteratively improves desirable behavior

---

## 🧪 Lab Highlight — Toxicity Classifier RM

You build a **binary classifier** to label model responses:
→ **Toxic** vs. **Non-toxic**

This RM is then used to push generation toward **non-harmful** completions.

Skills gained:

* Data labeling strategies
* Model evaluation for safety metrics
* How reward feedback refines policy

---

## 🔹 Alignment Challenges

| Challenge             | Description                                                  |
| --------------------- | ------------------------------------------------------------ |
| Over-alignment        | Too safe → refuses harmless queries                          |
| Reward hacking        | Model exploits loopholes to score high without being helpful |
| Bias introduction     | Human preferences may not be universal                       |
| Evaluation difficulty | Hard to quantify safety + ethics perfectly                   |

📌 Human feedback quality ≈ success of alignment

---

## 🧩 RLHF Effects on LLM Behavior

✔ More polite & safe
✔ Better reasoning through structured feedback
✔ Higher trustworthiness
✔ Stronger follow-instruction ability

⚠️ But still not perfect → hallucinations persist

This is why additional techniques like **RAG** are introduced in Week 4.

---

## 📌 Week 3 Core Takeaways

* Alignment ensures LLMs are **safe for real-world deployment**
* RLHF makes models:

  * **Helpful** (completes tasks well)
  * **Honest** (reduces hallucinations)
  * **Harmless** (prevents abuse)
* Reward models are crucial for **behavior shaping**
* Lab teaches how to classify toxicity to guide generation
