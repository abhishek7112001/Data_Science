```python
# Numpy
    Numerical Python Operations
    
# Properties:

    n-dimensional array provides fast array oriented arithmetic operations.
    1. It is a collection of homogeneous datatypes.
       ndarray >> n-dimensional array
    2. Numpy array is faster than list
    3. Array can be created using tuple/list
    
# Applications:

    1. Mathematical functions for fast operations on entire arrays of data
       without having to write loops.
    2. Tools for reading/writing array data to disk.
    3. Linear algebra, random number generation, and Fourier transform capabilities.

    1. Linear Algebra
    2. Statistics
    3. Probablity
    4. Trignometric (sin, cos, tan)
    5. Matrix
    6. Image Processing(cv2)
    7. Logrithmic

# Note:
    -While NumPy by itself does not provide functionality.
    -Understanding of NumPy arrays and array-oriented computing
    will help you use tools with array computing semantics.
    like pandas, much more effectively
    
    
    
    Ex:
        import numpy as np
        np.array([1,2,3])
        np.array((1,2,3))
        
    
# Homogeneous datatypes:

Def:
    Homogeneous means same type of data.

    int, float : convert int >> float
    int, float, complex : convert int ,float >> complex
    int, float, complex, str: convert int ,float ,complex >> str
    ( int >> float >> complex >> str )
        
        
# Change Datatype of an array:

    int >> float
    float >> int
    int >> complex
    float >> complex
    complex >> int  #invalid
    complex >> float #invalid

    x = np.array([2,3,4,5,6])

    x = np.array([2,3,4,5,6], dtype = 'float')
    

# Dimensions in array:

    200       >> 0-D
    [200]     >> 1-D
    [[200]]   >> 2-D
    [[[200]]] >> 3-D

# 1. 0-D array:

    Def:
        Zero dimension is an scalar.
        Can't access it via indexing.
    Ex:
        x = np.array(100)
        print(x.ndim)     # o/p >> 0

# 2. 1-D array:

    Def:
        It is collection of multiple 0 D arrays.
    Ex:
        x = np.array([200, 300, 400, 500], dtype = 'float')
        print(x.ndim)     # o/p >> 1
    
# 3. 2-D array:

    Def:
        It is collection of multiple 1 D arrays.
    Ex:
        x = np.array([[10,20],
                      [30,40]], dtype = 'float')
        print('Number od dimensions of an array : ', x.ndim)  # o/p >> 2
    
# 3-D array:

    Def:
        It is collection of multiple 2 D arrays.
    Ex:    
        x = np.array([[[2,3,4],[4,5,6],[6,7,8]], 
                      [[0,9,8],[5,4,3],[3,2,1]]])
        print('Number od dimensions of an array : ', x.ndim)
    
# Indexing and Slicing in numpy:

# 1d-array slicing:

    Syntax:
        [start_index : last_index : Stepsize]
    Ex:
        arr1 = np.array([10,20,30,40,50])
        arr1[2:4]
    
# 2d-array slicing:

    Syntax:
        [rows , columns ] == [start_index_row : end_index_row, 
                              start_index_col : end_index_col]
    Ex:
        array2 = np.array([[2,3,4],
                           [5,6,7],
                           [10,20,30],
                           [100,200,300]])
        array2[1][2] 
        # o/p 7

        array2[3][2]
        #o/p 300

        # [Row,Col]
        array2[3,2]
        #o/p 300

        # arr [start_index_of_rows : end_index_of_rows, start_index_of_col : end_index_of_col]
        arr[: , :] 
    
    
# 3D array indexing and slicing:

    arr = np.array([[[2,3,4],
                     [5,6,7],
                     [10,20,30]],
                    [[20,30,40],
                     [50,60,70],
                     [100,200,300]]])
    

# arr [2D][Rows][Col]
  arr[1][1][0]

# arr[ Plane, Rows, Col]
arr[1,1,0]

''' [start_index_of_2d   : end_index_of_2D,
     start_index_of_rows : end_index_of_rows, 
     start_index_of_col  : end_index_of_col ]
'''
arr[:,:,:]

```
