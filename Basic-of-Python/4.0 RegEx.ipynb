### `Regular Expression~`

```python
Def:
    A Regular Expression (RegEx) is a sequence of characters that defines a search pattern.

Properties:
    - It can detect the presence or absence of a text by matching it with a particular pattern,
      and also can split a pattern into one or more sub-patterns  
    - Python provides a re module that supports the use of regex in Python.
    - There are other several functions defined in the re module to work with RegEx    
   
# MetaCharacters:

        Metacharacters are characters that are interpreted in a special way by a RegEx engine. 
        Here's a list of metacharacters:
      
        [] . ^ $ * + ? {} () \ |


# functions: 

      Function    Returns
    1. findall >>  list
    2. search  >>  object
    3. match   >>  object
    4. sub     >>  str
    5. compile >>  object
    
# 1. findall
    Def:
        It returns list of substring of matches. 
    Syntax:
        re.findall('pattern', text)
        
# 2. search
    Def:
        It searches for a match and return an object of first occurrence only.
    Syntax:
        re.search(pattern,text)
    
# 3. match

    Def:
        It searches for a match at beginning of string  and return an object.
    Syntax:
        re.match(pattern,text)
 

# 4. compile

    Compile a regular expression pattern, returning a Pattern object.
    Syntax:
        p = re.compile(pattern)
        r = p.search(string)
        r.group()
        
# 5.sub

    Def:
        Same as replace function from string.
        Returns String
    Syntax:
        new_text = re.sub(pattern, new_string, text)
```

```python
# special sequences:

1. \d >> [0-9]

2. \D >> except [0-9] or [^0-9]

3. \w >> [a-zA-z0-9] 

4. \W >> except [a-zA-z0-9] or [^a-zA-z0-9] >> space, special ,escape

5. \s >> space

6. \S >> except space

7. \b >> char at start and end word

8. \z >>  Matches specified characters are at the end of a string
```

## `Pytesseract~`

```python
OCR = Optical Character Recognition.

Def:
    OCR systems transform a two-dimensional image of text to text i.e String.
    PDF(Protected Data File) >> imgae >> text
    
# Platforms:

    1. Pytesseract
    2. Google Vision
    3. Amazon Textract

# History:

    1. Pytesseract began as a Ph.D. research project in HP Labs, Bristol. 
    2. It gained popularity and was developed by HP between 1984 and 1994. 
    3. In 2005 HP released Tesseract as an open-source software. 
    4. Since 2006 it is developed by Google.
    5. Tesseract is an open source text recognition (OCR) Engine
    
    
# Properties:

    1. Data may contain machine printed or handwritten text. 
    from its image representation into machine-readable text.
    
    2. OCR as a process generally consists of several sub-processes i.e hirearchy
    
    Hirearchy:
            Block >> Paragraph >> Line >> Words
    
    
# Advantages:

    1.Tesseract is compatible with many programming languages.
    2.Detect in multiple languages.
    3.Tesseract performs well when document images follow the next guidelines:
        Clean segmentation of the foreground text from background
        Horizontally aligned and scaled appropriately
        High-quality image without blurriness and noise
        
    4.For almost two decades, optical character recognition systems have been
    widely used to provide automated text entry into computerized systems.
    
    
# Limitations of Tesseract

    1. It is not capable of recognizing handwriting.
    2. Doesn't do well with affected images or low quality images. 
    3. Poor quality scans may produce poor quality OCR.
    4. It does not expose information about what font family text belongs to.

        
# Preprocessing for Tesseract:

    1. To avoid all the ways your tesseract output accuracy can drop.
    you need to make sure the image is appropriately pre-processed.
    
    2. This includes rescaling, binarization, noise removal, deskewing, etc.
    To preprocess image for OCR, use functions from OpenCV.
```
