### `( lambda map reduce functions~ )`

```python
# Types of ...

    1. lambda arguements :  expression      
    2. map(function_name,iterable)
    3. filter(function_name,iterable)

# 1. lambda():

    # Lambda Properties:   
        >> Lambda keyword is used to define annonymous function
        >> annonymous : function without name

        >> defined by using lambda keyword
        >> No need to use return statement

        >> The expression is evaluated and returned
        >> Short period of function
        >> Higher - order function : If function is taking function as an arguemnt

    # Syntax:

        lambda arguements :  expression   
        ~ arguements >> Single or Multiple
        ~ expression >> only one expression

    # Use of Lambda Function in python
        ~ We use lambda functions when we require a nameless function 
        for a short period of time

# 1. map():

    # Properties:
        
        The map() function applies a given function to each item of 
        an iterable (list, tuple etc.) and returns an iterator.

    # Syntax:
    
        map(function, iterable,...)
        ~ function - a function that perform some action to each element of an iterable
        ~ iterable - an iterable like sets, lists, tuples, etc
        
    # Returns:
        >> The map() function returns an object of map class. 
        >> The returned value can be passed to functions like
            list() - to convert to list
            set()  - to convert to a set, and so on.
        
    # example:
    
            m=1
            def mul(a):
                global m
                m*=a
                return m
            x=list(map(mul,range(1,6)))[-1]
            x
            
            
# 2. filter():

    # Properties:
        The filter() function extracts elements from an iterable (list, tuple etc.)
        for which a function returns True.
        
    # Syntax:
        filter(function, iterable)
        ~ function - a function
        ~ iterable - an iterable like sets, lists, tuples etc.
        
    # Returns:
        The filter() function returns an iterator.
        ~ You can easily convert iterators to sequences like lists, tuples, strings etc
    
    # example:
        num = [1, 2, 3, 4, 5, 6, 7]
        even_num = filter(lambda x: x%2 == 0, num)
        even_num = list(even_num)
        print(even_num)

        
# 3. reduce():

    # Properties:
        -Cummulative way
        The reduce() function in python takes a pre-defined function and 
        apply it to all the elements in an iterable (eg: list, tuple, dictionary etc.)
        and compute single valued result.
        
    # Syntax:
    
        import functools
        functools.reduce(function, iterable)

        ~  This function will be applied to all the elements in an iterable in a 
           cumulative manner to compute result.
        ~  Iterables are those python objects that can be iterated/looped over, 
           includes like lists, tuples, sets, dictionaries, generators, iterator etc.
        ~  You can also pass 2 iterables in reduce     
    
    # Returns:
        The outcome of the reduce function is a bsically a sigle value return.

    
# difference between lambda, map & reduce:

    1. If we use Python functions map( ), filter( ), reduce( ) we do not need
        a for loop or if statement to control the flow. 

    2. This lets the programmer focus on the actual computation rather than on the
        details of loops, branches, and control flow.
```

## 3. `File Handeling~`

```python
# Files

Def:
    File handeling is used to create,write and read the files.
        
    Properties:
        Files are named locations on disk to store related information. 
        They are used to permanently store data in a non-volatile memory (e.g. hard disk).
        Since Random Access Memory (RAM) is volatile (which loses its data when the computer is turned off),
        we use files for future use of the data by permanently storing them.
        When we want to read from or write to a file, we need to open it first.
        When we are done, it needs to be closed so that the resources that are tied with the file are freed.
        Hence, in Python, a file operation takes place in the following order:
    
    File Operations~
        1. Open a file
        2. Read or write or append (perform operation)
        3. Close the file
```

## 5. `Exceptions~`

```python
# Properties:
    
Def:
    While creating and executing a Python program things may go
    wrong at two different stagesͶduring compilation and during execution.
     1. Errors that occur during compilation are called Syntax Errors.
     2. Errors that occur during execution are called Exceptions.
        
    Something in the program is not as per language grammar
    Reported by - Interpreter/Compiler
    Action to be taken - Rectify program
    
# Why to use?
    when we are not sure, if the code would yield desired results or not!
    
# Other common syntax error are:
    
    - Leaving out a symbol, such as a colon, comma or brackets
    - Misspelling a keyword
    - Incorrect indentation
    - Empty if, else, while, for, function, class, method
    - Missing :- Incorrect number of positional arguments
    
# Types of Error:

    - NameError
    - AttributeError
    - SyntaxError
    - IndentationError
    - IndexError
    - ZeroDivisionError
    - FileExistsError
    - TypeError
    - ValueError    
    print(dir(locals()['__builtins__'])) # to print all the errors 
    
# Python facilitates exception handling by providing:

    - Keywords:
            try, except, else, finally, raise.
    - Readymade exception classes.    
    
# How to deal with Exceptions?

    1. try and except blocks are used to deal with an exception.
    2. Statement(s) which you suspect may go wrong at runtime should be
       enclosed within a try block.
    (a) Catch the raised exception object in except block.
    (b) Raise the exception further.
 
# How to Use try-except:

    # builtin keywords used for error handeling
    >>
        1. try
        2. except
        3. raise
        4. finally
        5. else    
   >>

    (a) try    block - Enclose in it the code that you anticipate will cause an exception.
    (b) except block - Catch the raised exception in it. It must immediately
    
    try    - it test the block of code & check if it works
    except - if try doesnt work then except block will handle 
             the errors occuring in try block


    try:
        here we test the code
        (if the code runs successfully then except block will not be executed)

    except:
        in case condition or code in try block doesnt work 
        then this will handle the errors occuring in try block
        (we can print the errors here using traceback or even pass on the errors)
    
# basic example:
    try:
       # do something
       pass

    except ValueError:
       # handle ValueError exception
       pass

    except (TypeError, ZeroDivisionError):
       # handle multiple exceptions
       # TypeError and ZeroDivisionError
       pass

    except:
       # handle all other exceptions
       pass

# Examples:

    # 1. try-except
        var = (3,2,4,5,6)
        try:
            var.append(100)

        except:
            print('Error in try block')
        
    # 2. try-except 
        try:
            with open('sample.txt', 'r') as f:
            text = f.read()
            print(text)
        except:
            print('File not exists')

    # 3.try-except-traceback
    
       # traceback:
            A list of the functions that are executing, printed when an exception occurs.
            When we don't know the error then we use traceback method.
            
        import traceback
        try:
            tuple1 = (10,20,30,40)
            print(tuple1[5]) #Error
            print('Testing Try block')
        except:
            print(traceback.print_exc())
            print('except block')
        
    # 4. try-except-except
    
    text = 'Machine learning is a branch of artificial intelligence (AI) and computer science which'
    try:
        with open('data.txt','w') as f:
        f.write(text)
        print('File created')

        with open('data1.txt','r') as f: #Error
        text = f.read()
        print(text)

    except FileExistsError as err:
        print(err)
        print('Except block')

    except FileNotFoundError as err:
        print(err)
        print('Except block')
        
# nested try-except-finally

    try:
      f = open("demofile.txt")
      try:
        f.write("Lorum Ipsum")
    except:
        print("Something went wrong when writing to the file")
    finally:
        f.close()
    except:
      print("Something went wrong when opening the file")

```

### `iterator~` `generator~` ` decorator~`

```python
# Iterator:

Def:
    An iterator is an object whcih implements the iterators protocol
    which consist __iter__ and __next__ method
    
Properties:
     For loop works like an iterator
     iterators : iterator
     iterables : string, list, tuple, set, dict, range
     It contains __iter__ and __next__ methods            # iter and next function
    
    
# Example:

    l = [1,2,3]         # iterable 
    i = iter(l)         # returns iterator object
    print(next(i))      # prints first element
    print(next(i))      # prints second element
    print(next(i))      # prints third element
    print(next(i))      # returns StopIteration eroor...

# Generator

    >> Basically used to create an iterator
    >> If we are using one or more yield statements, then it is called as generator
    
# Example:

def my_fun(list1):
    for i in list1:
        sqr = i ** 2
        print(f'Square of {i} is {sqr}')
        yield sqr       # generator  >> It will make the entire function (my_fun) as iterator
    
my_fun([4,5,6,7])


# Difference between Iterator and Generator:

>> To create an iterator we use iter() function.    To create generator we use function 
                                                    along with yield keyword
                                                    Generator uses the yield keyword. 
                                                    It saves and return the variable value 
>> It returns iterator object.                      It returns generator object.          
    
    Every iterator is not a generator              Every generator is an iterator
    
 
# hirarchy:
    Iterable >> Iterator >> Generator
```

```python
# decorators

Def:
    Basically, a decorator takes in a function, 
    adds some functionality and returns it.(another function)
    In Decorators, functions are taken as the argument into another function 
    and then called inside the function.

# Example:

def outer_function():
    print('Outer Fucntion')
    
    def inner_function():
        print('Inner function')
        return 'PYTHON'
    
    return inner_function       # decorator

x = outer_function() # object
x()                  # inner_function


# Difference between ...

    1. What are iterator,generator,decorator.
    2. iterator vs generator
    3. print() & return
    4. return & yield
```
