```python
#  What's Oops?

Def : 
    It's an approach to solve a programming problem by creating an object.
    
    # Class: 
        Blueprint of an object that defines variables & methods(function).
        
    # Object:
        It's nothing but creating an instance of class.
        
    # Init:
        Init means initialization method (__init__)
        It's always called when an object is created.
        
    # Self:
        Self parameter is a reference to the current instance of class.
        Whenever we call instance method, address of the object get passed to
        the method implicitly. 
        This address is collected by the instance method in a variable self.
        
    # Super:
        super() is a build in method that returns an object (temporary obj of superclass)
        that allow us to access methods of baseclass.
        
        
# Difference between Object & Class:
    
           Class                    Object
   1. A class has a name       Objects are nameless
   2. Generic in nature        Specific in nature
   3. No memory is allocated   Memory is allocated
   4. Declared only once       Can create many times
   5. Can't manipulate class   Can manipulate object ( Due to memory allocation )
   6. Logical entity           Physical entity

# Types of variables in oops:

    1. Global Variable
    2. Local  Variable
    3. Class Variable:
           Public  Class Variable
           Private Class Variable
    4. Instance variable:
           Public  Instance Variable
           Private Instance Variable
    
# features of oops:
    There are five features of oops
    
    1. Classes & Objects
    2. Inheritance
    3. Encapsulation
    4. Polymorphism
    5. Abstraction
    
    
    
# 1. Classes & Objects:
    
# 2. Inheritance:

Def :
    It allows a class to inherit the properties from other class. 
    Creating a new class for using details of an existing class without modifying it.
        
    # Base Class / Parent Class / Super class :
        Class from which properties are being inherited
        
    # Derived class / Child class / Sub class :
        Class in which properties are inherited
        
    # Types of Inheritance:
    
        There are five types of inheritance.

        1. Single inheritance
        2. Multiple inheritance
        3. Multilevel inheritance
        4. Hirarchical inheritance
        5. Hybrid inheritance

        
        
# 1. Single Inheritance:
        
Def:
    When a child class inherits from only one parent class.
                
Example:
    class Parent():        # defining class
        pass
            
    class Child(Parent):   # inheriting Parent class
        pass 
            
    c1 = Child()           # created an object c1

        
        
        
# 2. Multilevel inheritance:
        
Def:
    When we have a child and grandchild relationship.
                
Example: 
    
    class GrandFather():        # Superclass
        print("Grandfather")
                    
    class Father(GrandFather):  # Father  is inheriting GrandFather
        print("Father")
                    
    class Child(Father):        # child  is inheriting Father
        print("Child >>")
                    
    obj=Father()                # creating an Object of class Child()
                
             
                
# 3. Multiple Inheritance:
        
Def:
    When a child class inherits from multiple parent classes.
                
Example:   
    
    class Father():
        print("Father")

    class Mother():
        print("Mother")

    class Child(Mother,Father):  # Multiple Inheritance
        print("Child")

    c = Child()
        
        
        
# 4. Hirarchical inheritance:

Def:
    More than one derived classes are created from a single base.
        
Example:
    class Parent():
        pass

    class Child1(Parent):
        pass

    class Child2(Parent):
        pass

    c1 = Child1()
    c2 = Child2()
        
        
# 5. Hybrid inheritance:

Def:
    It is a blend of more than one type of inheritance.
        
Example:
    class A:         # superclass/baseclass
        pass

    class B(A):
        pass

    class C(A):
        pass

    class D(B,C):    # inheriting B & C simultaneously
        pass

    d1 = D()
            
# 3. Encapsulation:

Def:
    Protecting from accidental modification of data. 
    Used to restrict the access of methods and variables from outside the class
    and  from other classes as well (child class)
        
Example:
    var0 = 'VB'                 # Global variable
    class A:
            
        var1    = 'Hello World'      # public class variable
        __var2 = 'Data Science'      # private class variable

    def __A1(self):             # private method
        var3        = 'Local'       # local variable
        self.var4   = 'Hi'          # public instance variable
        self.__var5 = 'LOL'         # private instance variable

    obj = A()   # object
    obj.__var2  # not accessible
    obj.__A1()  # not accessible
                
        
# Name mangeling:

Def: 
    Name mangeling is the process of to access the private var & methods.
    
Example:
    ar0 = 'VB'                  # Global variable
    class A:

    var1   = 'Hello World'   # public class variable
    __var2 = 'Data Science'  # private class variable

    def __A1(self):             # private method
    var3        = 'Local'       # local variable
    self.var4   = 'Hi'          # public Instance variable
    self.__var5 = 'LOL'         # private instance variable

    obj = A()       #  object
    obj._A__var2    #  private variable accessible
    obj._A__A1()    #  private method accessible
                    
                

        
# 4. Polymorphism:

Def:
    means the same function name being used for different types.
    
Example:
    
    class Circle():
        def area(self):
            print('Area of Circle')

    class Square():
        def area(self):
            print('Area of square')

    class Rectangle():
        def area(self):
            print('Area of rectangle')

    c1 = Circle()
    s1 = Square()
    r1 = Rectangle()

for Obj in [c1,s1,r1]: # c1
    Obj.area()         # c1.area(10)
                       # s1.area()
                       # r1.area()
    
    
    
# 5. Abstraction:

Properties:
    It is used to hide internal functionality of the function.
    Blueprint of the project.
    Abstract method is used for declaration only.
    Does not used for implementation.

Example:
    from abc import ABC,abstractmethod

    class ProjectBlueprint(ABC):        # Abstract Class
        @abstractmethod                 # decoractor
        def registration(self):
             pass

        @abstractmethod                 # abstract method
        def login(self):
            pass

        @abstractmethod
        def forget_password(self):
            pass

        def process_document(self):    # Normal method
            pass
    

        
```
