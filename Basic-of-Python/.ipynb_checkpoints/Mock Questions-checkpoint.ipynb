## 2. `Linear Regression`


```python

 - Aim:

    It is predective model used to find the linear relationship between 
    dependent variable (Y) and one or more independent variable  (X).
    Based on BFL
    
    Linear        >> straight line or straight path
    regression    >> Predicting continous values or real number
    Target column >> (Continous values)Prices,Sales,Marks,Temp,Age, Cibil Score etc.
    
# - Types of Linear regression:
    
1. Simple Linear regression:
    
    Only one independent variable
        Y =  mx +c
        m >> slope
        c >> intercept
        x >  Independent variable
        y >> dependent variable e.g score,Price etc
        
2. Multiple Linear regression:
    
    Two or more independent variable
        Y = m1x1 + m2x2 + m3x3.....+mnxn + c
        m1,m2,m3   >> slopes
        x1,x2,x3   >> Independent variables
        Y          >> dependent variable
        
# - Assumptions of Linear regression (Explain each in details):

1. Linearity:
    To assume the linearity of data
    Linear relationship between independent variable and dependent variable.
    
    Co-efficient of correlation(R - Value):
        
    Def: 
        It tells us Strength & direction of relationship.

    Formula:      
                    Covariance        
    Correlation = ------------------- 
                   product of std dev.
        Std dev = tells us hoe the data is spreaded.
        Variance = Tells us how the data is deviated from mean value.
                  

    Range for R value:   : -1 to +1 
    Positive Correlation : R = +1
    Negative Correlation : R = -1
    No Correlation       : R = 0
    Good Predictors      : R >= 0.7 
    Bad Predictors       : -0.3 < R < 0.3 

        
    2. Scatter Plot       : sns.pairplot(df)
    3. Correlation matrix : 
        
2. No Multicolineartity:
    All the variables/features should be independent to each other.
    
    VIF (Variable Inflation Factor) :
        Corelation of independent features.
        
    from statsmodels.stats.outliers_influence import variance_inflation_factor 

    Formula:
                1
        VIF = ------
              (1-R^2)
                
        Ideal Range : 1  to 10 
        It can go   : 1 to  infinity
        VIF         = 1 → No correlation
        VIF         = 1 to 5 → Moderate correlation
        VIF         >10 → High correlation
        
        
    
3. Normality of residuals (error):
    There are five methods to check whether the residuals are normally distributed or not.
    Two methods are for plotting techniques
    Three methods are determined using p-Value
    Out of those five methods any three becomes true then we can say data is normally distributed
    
    1. Plotting Techniques:
        import statsmodels.api as sm
        KDE PLOT : sns.kdeplot( residual_train, fill=True )
        QQ-PLOT  : sm.qqplot(residual_train, line = '45', fit = True)
            
    2. Hypothesis Techniques:
        
        # Libraries & functions
        from scipy.stats import shapiro,kstest,normaltest
        
        Shapiro test
        KS Test
        Normal Test
        
        In the above three tests we calculate the p-Value.
        p-Value >= 0.05 - We are accepting Null Hypothesis
        p-Value < 0.5  - We are accepting Alternate Hypothesis
        
        Null Hypothesis      : states that the data is normally distributed
        Alternate Hypothesis : states that the data is not normally distributed
    
4. Homoscedasticity:
    It means constant error between Ya-Yp.
    It means having same scatterplot.
    The data points must be about the same distance from the line.
    If the outliers are present we will not get homoscedasticity.
    If the outliers are not present we will get homoscedasticity.
    
    sns.scatterplot( x = y_train, y = residual_train)
    
```

### 5. `Evaluation Metrics for regression`

```python
# 1. Residual

    Def:
        Residual means Error.
        Error between Actual values and Predicted values
    Formula:
        Residual = (Yactual - Ypred)

    Positive >> If the data points are above the regression line
    Negative >> If the data points are below the regression line
    zero     >> If the data points are on the regression line
    
 # 2. Sum of Squared Error
    Def:
        Sum of squared difference between actual values and predicted values.
        To lower the impact of negative values.

    Formula:
        SSE = sum(Ya - Yp)^2

    
# 3. Sum of squares due to regression
    Def:
        Sum of squares due to regression (SSR)
        It is explained variation.
    Formula:
        SSR = sum(Yp - Ymean)^2
        
    SSR >> Sum of suared difference between predicted values and
           mean value of dependent of variable

# 4. Total Error
    Def:
        It is un-explained variation.
    Total Error ( SST)
    SST = sum(Ya - Ymean)^2
    SST >> Sum of squared difference between Actual values and
           mean value of dependent variable

# 5. Mean Squared Error (Cost function):

    Every algorithm has its own Cost function.
    For linear regression it is a MSE.
    Our Aim is to reduce the MSE.
    Less MSE < Accuracy is high.
    To reduce Cost Function we use GD Algorithm.
    BFL is find out by using Cost function.

    MSE = (sum(Ya - Yp)^2) /N 
    MSE >> mean of Sum of squared difference between actual values and predicted values
           mean of residuals.

# 6. R-Squared:
    Coefficient of Determination
    Def:
        It tells us goodness of BFL.
        
    FORMULA:
        R-Squared = 1 - (SSE/SST)

    R-Squared Value:

        >> Coefficeint of Determination
        >> Scale invarient  #Not affected by scale
        >> 0 to 1
        >> It can be Negative
        >> Goodness of BFL
        >> Ideally value of R^2 should be lower.

# 7. Adjusted R-Squared:
    Ideally value of R^2 should be lower.
    When we have more non-corelated features R^2 increases.
    To overcome this drawback we use Adjusted R-Squared Value.
    
    FORMULA:
        _
        R^2 = 1 - (1-R^2) * (n-1) / (n-P-1)
        
        n = no of samples.
        P = no of predictors.
    
```

### 9. `Overfitting, underfitting`

```python

BIAS and VARIANCE are the terms used to explain the accuracy of model.


# Bias :

    Def:
        It is considered for Training Data.
        
    LOW  BIAS = High Training Accuracy
    HIGH BIAS = Low  Training Accuracy
    
    Ex:
        Training  Accuracy = 96% -LOW BIAS
        Training  Accuracy = 70% -HIGH BIAS
        
    
# Variance :

    Def:
        Difference between accuracies. 
        (Training Accuracy - Testing Accuracy)
        
    Low TRAINING & High TESTING  = High Variance
    High TRAINING & Low TESTING  = High Variance
    Low TRAINING & Low TESTING   = LOW Variance  # Best Model
    High TRAINING & High TESTING = LOW Variance  # Best Model
    
    
# - Best model in terms of bias and variance:

    If the Model is performing well on TRAINING data as well as TESTING data.
    
    Training  Accuracy = 96% }
                              } # Low Bias and Low Variance
    Testing  Accuracy  = 95% } 
    
# - Techniques to reduce the impact of overfitting:

    Def:
        Regularization increases Bias a little bit,
        but in return we get a big decrease in Variance.

    Regularization technique is used to overcome overfitting.
    It includes L1 & L2 Regression.
    i.e Lasso & Ridge Regression.
    By increasing the Bias a little, we can decrease the Variance.
    
    
- Techniques to reduce the impact of underfitting

    Increase the dimensionality of model.
    Increase the complexity.
    Increasing the training time until cost function in ML is minimised.
    
```

```python
# 18.What is confusion metrics and explain:

    Confusion Matrix is used to find out the accuarcy of LR.
    It tells us the proportion of both +ve & -ve Classes.
    
    
# - Precision formula and explain with example & application:

    Def:
        Out of total predicted positive values, 
        how many are actual positive (correctly classified) 
    FORMULA:
                            TP
            Precision = -----------
                         (TP + FP)
    
    
# - Recall formula and explain with example & application
    Def:
        
    FORMULA:

                        TP 
            Recall = ----------
                     (TP + FN)

    >> Importance to FN
    >> FN should be less
       - TP,FP,FN,TN
        TP = True Positive  - Correctly Classified data points for Class-1
        TN = True Negative  - Correctly Classified data points for Class-2
        FP = False Positive - Incorrectly Classified data points for Class-2
        FN = False Negative - Incorrectly Classified data points for Class-1
        
#  TPR :
    TPR >> True positive rate   >> Sensitivity  >> Recall

    TPR = TP/ (TP+FN)

    Proportion of positive class got correctly classified by classifier
    
    
# - FPR:
    FPR >> False Positive Rate

    FPR = FP/(TN + FP)

    Proportion of negative class got incorrectly classified by classifier


# - TNR:
    TNR >> True Negative rate >> Specificity >> 

    TNR = TN / (TN + FP)
    Proportion of negative class got correctly classified by classifier
    
    
# - FNR:
    FNR >> False Negative Rate

    FNR = FN / (TP+FN)

    Proportion of positive class got incorrectly classified by classifier
    
    
```

```python
# 19.What is type-1 error and Type -2 error

Type-1 error:
    Incorrectly classified data points for class 2
    A type I error occurs when the null hypothesis is true but it is rejected.
    
Type-2 error = Incorrectly classified data points for class 1.


Type I error:
    Rejecting the null hypothesis when the null hypothesis is true.
    denoted by the Greek letter \alpha α (alpha).
    
Type II error:
    Accepting the null hypothesis when the null hypothesis is false.
    denoted by the Greek letter \beta β (beta)
    
```
