# `'String'`

```python
# What's string?

    1. Computers do not deal with characters, they deal with numbers (binary).
    2. Even though you may see characters on your screen, 
       internally it is stored and manipulated as a combination of 0s and 1s.
    3. This conversion of character to a number is called encoding,
       and the reverse process is decoding. 
    4. ASCII and Unicode are some of the popular encodings used.
    
# What's Unicode?

    -In Python, a string is a sequence of Unicode characters.
    -Unicode was introduced to include every character in all languages
    and bring uniformity in encoding. 
```

```python
# String:

Def:
    -A type that represents sequences of characters represented in 
    single,double or triple quotes (' ' or " " or """ """)

Properties:
    >> String is a collection of characters surrounded by 
       single quotes, double quotes or triple quotes.
    >> We can not use single quotes in single quotes
    >> We can not use double quotes in double quotes
    >> Strings are immutable  >> unchangable
    >> We cannot modify string once created
    >> concatenated by using + operator
    
Functions:

# returns Boolean value (True/False):

    >> isalnum()
        True  >> if all characters in the string are alphanumeric 
                (either alphabets or numbers)
        False >> If not, it returns False.

    >> isalpha()  
        True  >> If string contains only alphabets, a-z A-z 
        False >> It does not allow special char and space


    >> isdecimal()
       True  >> if all characters in the string are decimal characters.
       False >> if at least one character is not decimal character.
        # Unicode are allowed
        # Special chars,dots, spaces are not allowed
        
    >> isdigit()
        True  >> If the string contains digits, unicode,subscript & superscripts
        False >>It does not allow special chars (dot,space,etc)

    >>isupper()
        True  >> if all characters in a string are uppercase characters
        False >> if any characters in a string are lowercase characters

    >> islower()
        True  >> if all alphabets that exist in the string are lowercase alphabets.
        False >> if the string contains at least one uppercase alphabet.

    >> isnumeric()
        True >> If all char are digits, unicode, subscript, superscript, 
        vulgar fractions : (1/2, 2/3, 3/4)

    >> isspace()
        True  >> if string contains only spaces 
        False >> ' ab '.isspace()

    >> istitle()
        True >> if string is in title case or not
        False >> ' A b%* '.istitle()
        
    >> endswith()  
        True >> if the string is ending with a definite substring
             We can give user defined length of a string also.

    >> startswith() 
        True >> if the string is starting with a definite substring
             We can give user defined length of a string also.


# returns list

    >> split()    
        -Splits the string at the specified separator, and returns a list of substring

    >> splitlines()
        -Splits the string at line breaks and returns a list


# returns String

    >> join()     
        -Converts the elements of an iterable into a string

    >> capitalize()
        -Converts the first character to upper case

    >> lower()    
        -Converts whole string into lower case

    >> casefold()
        -Converts string into lower case
        -Convert german word (beta) into 'ss'

    >> replace()
        -Returns a string where a specified value is replaced with a specified value

    >> title()  
        -Converts the first character of each word to upper case

    >> upper() 
        -Converts whole string into upper case

    >> strip()   
        -Returns a trimmed version of the string 

    >> center() 
        -Returns a centered string with
        Even length = adds leading spaces 
        Odd length  = adds trailing spaces 

    >> rstrip()    
        -Returns a right trim version of the 
        
    >> lstrip()    
        -Returns a left trim version of the string


# returns Index

    >> index() 
        -Searches the string for a specified value and 
        returns the position of where it was found
        - returns error if the element not found
        We can give user defined length of a string also.
        

    >> find() 
        -Searches the string for a specified value and 
        returns the position of where it was found    
        - returns -1 if the element not found
        We can give user defined length of a string also.
        
        
        Index                    Vs             find
Returns index value if found          Returns index value if found 
Returns error if not found            Returns -1 if not found


# returns Number

    >> count()
            -Returns the number of times a specified value occurs in a string
             We can give user defined length of a string also.
             count( 'A' , start_index, end_index)

    
# makeup with zero

    >> zfill() 
            -Fills the string with a specified number of 0 values at the beginning only.
             Not vice a vesra.
    

# Built-in functions on string:
   
len max min chr ord sum print type 
```

## `Operators~`

```python
# Operators:
    -Operators are used to perform operations on variables and values.

Types of operators:
    1. Comparison Operators
    2. Logical Operators
    3. Arithmetic Operators
    4. Assignment Operators
    5. Membership Operator
    6. Identity Operators
    7. Bitwise Operator

    
# 1. Comparison Operators:

Def:
    -These operators compare the values on either sides of them and 
    decide the relation among them. 

    a == b         # Equal operator
    a != b         # Not equal
    a > b          # greater than
    a < b          # less than
    a >= b         # greater than or equal to
    a <= b         # less than or equal to

# 2. Logical Operators:

Def:
    -These are the special symbols that carry out arithmetic  and logical computations.
    -The value the operator operates on is known as Operand.

    1. and operator
    2. or operator
    3. not operator

# 3. Arithmetic Operators:

Def:
    -The arithmetic operators perform addition, subtraction,
     multiplication, division, exponentiation, and modulus operations

    (), **, %, *, /, //, +, -

    ()  >> Parenthesis
    **  >> Exponent
    %   >> Modulus # Remainder
    *   >> Multiplication
    /   >> Division
    //  >> Floor Division
    +   >> Addition
    -   >> Substraction

# 4. Assignment Operators:

Def:
     -Assignment operators are used to assigning value to a variable.

    =     >>   a=1
    +=    >>   a += 1   >>   a = a + 1 
    -=    >>   a -= 1   >>   a = a - 1
    *=    >>   a *= 2   >>   a = a * 2
    /=    >>   a /= 2   >>   a = a / 2
    //=   >>   a //= 2  >>   a = a // 2
    %=    >>   a %= 2   >>   a = a % 2
    **=   >>   a **= 2  >>   a = a ** 2

# 5. Membership Operator:

Def:
    -an operator that is used to validate the membership of a value

    1. in
    2. not in


# 6. Identity Operators:

Def:
    -Identity operators are used to compare the memory location of two objects.

    1. is 
    2. is not
    # -5 to 256 

# 7. Bitwise Operator:

Def:
    -Operators are used to perform operations on values and variables. 
These are the special symbols that carry out arithmetic and logical computations. 
The value the operator operates on is known as Operand. 

    &  = AND
    |  = OR
    ^  = XOR
    >> = RIGHT SHIFT
    << = LEFT SHIFT
    ~  = COMPLIMENT
```

# `Flow control Statements~`

```python
Def:
    Decision Making Statement
        
    - If else statements are basically decision making statements.
      (check for True or False)
    - If if is true then if will be executed & if if is false then else will be executed
    - If & Elif can not be empty but else can be empty(condition must be empty)


Syntax:
    if  condition:   
            Block of code
            statement 1
            statement 2
            ...
    else:
            statements
            ...

            
#e.g.

'''In this program, we input a number
check if the number is positive or
negative or zero and display
an appropriate message
This time we use nested if statement'''

num = float(input("Enter a number: "))
if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive number")
else:
    print("Negative number")
    
    
# valid statements:   

    >> if
    >> if >> else
    >> if >> elif
    >> if >> elif >> else
    >> if >> elif >> elif  >> else
    
    

# Nested if-else statements:
    -if there is if statement insde if then its called nested if- else.

if (condition1):
    if(condition2):
        statement
        ...
    else:
        statement
        ...
else:
    statement
    ...


# Short hand if else
Def:
    One line statement

    if: statement
    elif: statement
    else: statement
```

# `Types of loop~~`

```python
# types of loop
    1. for loop
    2. while loop
>>
1. for loop:
    
Def:
    The for loop in Python is used to iterate over a sequence
    (list, tuple, string) or other iterable objects. 
    Iterating over a sequence is called traversal.
    
Iterables:
    >> str, List, tuple, Set, Dict, Series, Range etc..

# Function:
    -is used to iterate over a sequence.
    -They basically execute the set of statements multiple times or 
     perform an operation on multiple items of the iterables.

# Syntax:
    for var in iterables(list,tuple,set...): # here var is iterator
        print(var*3)

# e.g.
    Syntax: range(start number, end number, step size)
    for i in range(10):  # Range creates a sequence of whole numbers 
        print(i)
```

```python
2. while loop:
    
Def:
    The while loop in Python is used to iterate over a block of code
    as long as the test expression (condition) is true.
    
Function:
    - We generally use this loop when we don't know the number of times
    to iterate beforehand.
    - Python interprets any non-zero value as True.
    - None and 0 are interpreted as False.
    
Syntax:
    while test_expression:
        Body of while   
    
# e.g.
'''Example to illustrate
the use of else statement
with the while loop'''

counter = 0

while counter < 3:
    print("Inside loop")
    counter = counter + 1
else:
    print("Inside else")
    
# how to take multiple inputs from users ?

while True :
    str1 = input("enter the text : ")
    if str1 =="stop":
        break
#     print(str1)
```

## `Loop control statements~`

```python
# Loop Control Statements:

Def:
    Loop control statements change execution from its normal sequence.
When execution leaves a scope, all automatic objects that were created
in that scope are destroyed. 


Functions:
    
    In Python, break and continue statements can alter the flow of a normal loop.
>>
    continue = The continue statement is used to skip the rest of the code inside
               a loop for the current iteration only. 
               Loop does not terminate but continues on with the next iteration..
>>    
    pass     = doesn't do anything on execution.
               when a statement is required syntactically but you do not want any 
               command or code to execute.
>>
    break    = Terminates the loop statement and transfers
               execution to the statement immediately following the loop.
```

# `{ Set }`

```python
# Properties:
    -A set is an unordered collection of items. 
    -Every set element is unique (no duplicates) and must be immutable (cannot be changed)
    -However, a set itself is mutable.
    -We can add or remove items from it.
    -Sets can also be used to perform mathematical set operations 
     like union, intersection, symmetric difference, etc.
>> set takes only immutable elements:( int, str, float, tuple, bool)

# Application:
    Basic uses include membership testing and eliminating duplicate entries.

 **Note**
    When creating set, be sure to not use empty curly braces {} or you will get an empty dictionary instead.
    
s = {1, 2, 3}
s = set([1, 2, 3])

s = {}  # this will create a dictionary instead of a set
s = set() # set
```

```python
# functions:

# returns boolean value:
    1. issubset()    # returns boolean value
    2. issuperset()  # Returns True if this set contains another sete (vice versa to subset)
    3. isdisjoint()  # Returns True if two sets have a null intersection

# returns new set 
    1. union()        # union() or | will create a new set that with
                      # all the elements from the sets provided.
    2. intersection() # intersection or & will return a set with only the
                      # elements that are common to all of them.
    
    3. differecnce()  # difference or - will return only the elements 
                      # that are unique to the first set (invoked set).
    4. symmetric_differecnce() # symetric_difference or ^ will return
                               # all the elements that are not common between them.

# updates old variable:

    1. add()      # We can add a single element to the set.
    2. update()   # And with update(), multiple immutable element.
    3. remove()   # Removes an element from the set. 
                  # takes ele as an arguement. returns error if ele is not present
    
    4. discard()  # discard() won’t raise any errors.
    5. pop()      # randomly removes element, Raises KeyError if the set is empty.
    6. clear()    # Removes all elements from the set
    
    7. intersection_update() # Updates the set with the intersection of itself and another.
    8. difference_update()   # returns Unique values from set1 and update set1
    9. symmetric_difference_update() # Unique values from both sets
```

```python
# Built-in functions with Set:

Built-in functions like all(), any(), enumerate(), len(), max(), min(),
sorted(), sum() etc. are commonly used with sets to perform different tasks.

>>

    all()       # Returns True if all elements of the set are true 
                # (or if the set is empty).
    any()       # Returns True if any element of the set is true. 
                # If the set is empty, returns False.

    enumerate() # Returns an enumerate object. 
                # It contains the index and value for all the items of the set as a pair.
    len()       # Returns the length (the number of items) in the set.

    max()       # Returns the largest item in the set.
    min()       # Returns the smallest item in the set.

    sorted()    # Returns a new sorted list from elements in the set
                # (does not sort the set itself).
    sum()       # Returns the sum of all elements in the set.
```

# `frozenset( {1,2,3...} )`


```python
# frozenset:

Properties:
    -Frozenset is a new class that has the characteristics of a set,
     but its elements cannot be changed once assigned.
    -While tuples are immutable lists, frozensets are immutable sets.
    -Sets being mutable are unhashable, so they can't be used as dictionary keys.
    -On the other hand, frozensets are hashable and can be used as keys to a dictionary.
    -Frozensets can be created using the frozenset() function.
  
# Methods:
    -This data type supports methods like copy(), difference(), intersection(),
    isdisjoint(), issubset(), issuperset(), symmetric_difference() and union(). 
    -Being immutable, it does not have methods that add or remove elements.

# frozenset function:
    
    > Unordered
    > Immutable
    >>Duplicates are not allowed
    
# All of the above functions returns frozenset.
    
    1. union()
    2. intersection()
    3. difference()
    
    4. symmetric_difference()
    5. issubset()
    6. issuperset()
    7. isdisjoint()
```

# { Dictionary : functions }

```python
Def:
    
Properties:    
    1. It is collection of key and value pair
    2. We cannot access items by using index
    3. Duplicate keys are not allowed, but values can be duplicate
    4. Dict are comma seperated.
    5. Enclosed by {}

Syntax:
    dict1  = {key1 : Value1, key2 : value2}
    Key    >> Immutable Datatypes (int, float, string, tuple, frozenset)
    values >> Any Datatype (int,float,string,list,tuple,set,dict, etc)
```

```python
# functions:>
    1. get()     # returns the value for the specified key if the key is in the dictionary.
    2. items()   # returns list of dictionary's (key, value) tuple pairs.
    3. keys()    # returns the list of keys as a view object
    
    4. values()  # returns the list of values 
    5. pop()     # method removes and returns an element from a dictionary having the given key.
    6. update()  # method updates the dictionary with the elements from another dictionary object or from an iterable of key/value pairs
    
    7. popitem() # Removes the last item inserted
    8. clear()   # removes all items from the dictionary.
    9. copy()    # returns a copy (shallow copy) of the dictionary.
    
    10. fromkeys()   # cretes dictionary with same value, 
                     # if value is not given >> 'k':None
    11. setdefault() #
    11. del()       # del can remove all the elements or it can remove specified key-pair
    12. zip()       # it zips the two sequence Datatypes 

#  **kwargs 
    >> Used to merge two dictionaries:
        
    d1={ 'Vishal' : 24, 'Vaibhav' : 21}
    d2={ 'VB' : 'Solapur', 'AB' : 'Pune'}
    d={**d1,**d2}
    print(d)
```
