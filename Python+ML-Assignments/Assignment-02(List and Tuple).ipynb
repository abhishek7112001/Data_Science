# 3.

append()
and
extend()
methods
for
lists.

The append() method adds a single object to the end of the list. Even if the item is another list, it is added as one entry, creating a nested structure.


```python
nums=[1,2,3]
nums.append([3,4])
print(nums)
```

The extend() method adds each item from an iterable (like a list, tuple, or string) individually to the end of the list. 

```python
nums=[1,2,3]
nums.extend([3,4])
print(nums)
```

# 25.
How
to
sort
a
list
in
ascending
and
descending
order
without
using
the
sort
function?

```python
# M-1 Bubble sort
lst = [78,7,4,3,9,0,5,6]
n= len(lst)
for i in range(n):
    for j in range(0, n-i-1):
        if lst[j]>lst[j+1]:
            lst[j], lst[j+1] = lst[j+1], lst[j]

print(lst)


```

```python
# M-2 Insertion sort
lst = [78,7,4,3,9,0,5,6]
n= len(lst)
for i in range(1, n):
    for j in range(i, 0,-1):
        if lst[j]<lst[j-1]:
            lst[j],lst[j-1] = lst[j-1], lst[j]
        else:
            break
print(lst)
```

```python
# M-3 Selection sort
lst = [78,7,4,3,9,0,5,6]
n= len(lst)
for i in range(n):
    mini_idx=i  
    for j in range(i+1, n):
        if lst[j]<lst[mini_idx]:
            mini_idx=j
    lst[i], lst[mini_idx]=lst[mini_idx], lst[i]

print(lst)
    
```

# 28.
Difference
between
copy
and
deep
copy?

```python
# copy: or shallow copy is a method that creates a list that copies all the elements of the original one
# but pointing to the same memory address

import copy
lst1= [[1,2], [3], 4, [5,6]]
shallow = copy.copy(lst1)
shallow[3][1]=100
print(lst1)
print(shallow)
```

```python
# deepcopy : This creates new list at different memory
import copy
lst1= [[1,2], [3], 4, [5,6]]
deep = copy.deepcopy(lst1)
deep[3][1]=100
print(lst1)
print(deep)
```

# 31.
Write

Python
program
to
convert
a
list
of
multiple
integers
into
a
single
integer
○
[11,
33,
50]
>>>
113350

```python
lst = [11,33,50]
n= len(lst)
if n==0:
    print(0)
else:
    ans=lst[0]
    for i in range(1,n):
        multiplier=1
        num=lst[i]
        while num>0:
            num//=10
            multiplier*=10
        ans= ans*multiplier+lst[i]
    print(ans)
```
