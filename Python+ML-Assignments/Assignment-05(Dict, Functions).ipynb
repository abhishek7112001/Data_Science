# 1.
Print anagrams together in python using a list and dictionary.
Input : arr = ['cat', 'dog', 'tac', 'god', 'act']
Output : 'cat tac act dog god'

```python
arr = ['cat', 'dog', 'tac', 'god', 'act']
dictio={}
for i in arr:
    new_i = "".join(sorted(i))
    if new_i not in dictio:
        dictio[new_i]=[]
    dictio[new_i].append(i)
res=[]
for group in dictio.values():
    res.extend(group)

print(" ".join(res))
```

# 2.
Python counter to find the size of the largest subset of anagram words.
Input ::> "qwe ewq rty ytr ytr ytr"
Output ::> 4
Input :: > "cars bikes arcs steer"
Output :: > 2

```python
arr = ['cat', 'dog', 'tac', 'god', 'act']
dictio={}
for i in arr:
    new_i = "".join(sorted(i))
    if new_i not in dictio:
        dictio[new_i]=[]
    dictio[new_i].append(i)
maxi=0
for group in dictio.values():
    maxi=max(maxi, len(group))

print(maxi)
```
