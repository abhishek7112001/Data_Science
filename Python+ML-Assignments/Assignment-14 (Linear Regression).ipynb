# 1.
What is Linear regression?

Linear Regression is a fundamental supervised learning algorithm used to model the relationship between a dependent variable and one or more independent variables. It predicts continuous values by fitting a straight line that best represents the data.

-It assumes that there is a linear relationship between the input and output
-Uses a best‑fit line to make predictions
-Commonly used in forecasting, trend analysis, and predictive modelling

# 4.
What are the assumptions of the Linear regression model?

1. Linearity: The relationship between the independent variable(s) and the dependent variable must be linear. This means that for every unit change in $x$, the change in $y$ is constant.
-How to check: Look at a scatter plot of $x$ versus $y$. If the data looks curved (e.g.,        exponential or parabolic), a simple linear regression will not be effective without transforming the data (e.g., using a log or polynomial transform).

2. Independence (No Autocorrelation): The observations (data points) should be independent of each other. This is especially critical in time-series data, where the value at time $t$ should not be predictable based on the value at time $t-1$.

3. Normality
The residuals (the differences between the observed values and the predicted values) should be normally distributed. This is required for calculating reliable confidence intervals and hypothesis testing.

How to check: Look at a Q-Q (Quantile-Quantile) plot or a histogram of the residuals. They should follow a straight line or a bell curve.

4. Equal Variance (Homoscedasticity): The variance of the residuals should be constant across all levels of the independent variable. 
How to check: If the spread of the residuals increases or decreases as $x$ increases, the model has "Heteroscedasticity."

# 5.
What if these assumptions get violated?

1. Linearity violation: It will result in under-fitting
2. Independence Violation: the model underestimates the standard errors. This leads to artificially small p-values, making you think variables are statistically significant when they might not be.
3. Normality violation: the statistical tests (t-tests, F-tests, confidence intervals) become invalid. You won't be able to trust your confidence intervals.

Normality
The residuals (the differences between the observed values and the predicted values) should be normally distributed. This is required for calculating reliable confidence intervals and hypothesis testing.

# 9.
What does multicollinearity mean?

-Multicollinearity occurs when two or more independent variables depends on each other.
-It means each variables are telling same story. Thus, the model struggles to determine the most  specific impact of each variable on target.


# 10.
How to check Multicollinearity?

```python
The most common way to detect this is using the Variance Inflation Factor (VIF).

The VIF measures how much the variance of an estimated regression coefficient increases if your predictors are correlated.

VIF = 1: No correlation.

VIF = 1 to 5: Moderate correlation (usually acceptable).

VIF > 5 or 10: High correlation (indicates significant multicollinearity that should be addressed).
```
