# 65.
python program to find the minimum sum of factors of a number.
Input : 12
Output : 7
Explanation:
Following are different ways to factorize 12 and sum of factors in different ways.
12 = 12 * 1 = 12 + 1 = 13
12 = 2 * 6 = 2 + 6 = 8
12 = 3 * 4 = 3 + 4 = 7
12 = 2 * 2 * 3 = 2 + 2 + 3 = 7
Therefore minimum sum is 7
