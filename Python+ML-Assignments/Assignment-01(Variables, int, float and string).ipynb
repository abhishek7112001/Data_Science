# 6.
What
do
you
mean
by
Python
literals?

In Python, literals are notations that represent fixed values directly in your source code. 
They are used to denote specific data types and values. Here are some common types of literals in Python:
1. String Literals: Enclosed in single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).
   Example: 'Hello', "World", '''This is a string''', """Another string"""
2. Numeric Literals: Represent numbers and can be of various types such as integers, floating-point numbers, and complex numbers.
   Example: 42 (integer), 3.14 (floating-point), 1 + 2j (complex)
3. Boolean Literals: Represent the two boolean values, True and False.
   Example: True, False
4. None Literal: Represents the absence of a value or a null value.
   Example: None
5. Collection Literals: Represent collections of data, such as lists, tuples, sets, and dictionaries.
   Example: [1, 2, 3] (list), (1, 2, 3) (tuple), {1, 2, 3} (set), {'key': 'value'} (dictionary)
6. Special Literals: Python also has special literals like Ellipsis (...) and NotImplemented, which are used in specific contexts.
   Example: ... (Ellipsis), NotImplemented

# 13.
W rite
a
Python
program
to
count
the
number
of
vowels,
consonants,
digits
and
special
characters
in
a
string

```python
vowels = consonants = digits = special =0
sentence= "This is 12orange3 juice&"
sentence = sentence.lower()
for i in sentence:
    if i in "aeiou":
        vowels+=1
    elif i.isalpha():
        consonants+=1
    elif i.isdigit():
        digits+=1
    elif not i.isspace():
        special+=1
    
print(vowels, digits, consonants, special)
```
