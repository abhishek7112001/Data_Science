# 1.
What
are
loop
control
statements
in
Python?

1. break: Immediately terminates the current loop and resumes execution at the next statement after the loop.
2. continue: Skips the remaining code in the current iteration and jumps back to the beginning of the loop for the next iteration.
3. pass: A null operation that does absolutely nothing when executed.
   Use Case: Acts as a syntactic placeholder for future code or to create an empty loop structure without causing an error.

# 28.

```python
n=int(input())

row=[1]
for i in range(0,n):
    print(row)
    curr_row=[1]
    for j in range(len(row)-1):
        curr_row.append(row[j]+row[j+1])
    curr_row.append(1)
    row=curr_row

```
