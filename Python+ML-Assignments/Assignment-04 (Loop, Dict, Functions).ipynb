# 7.
Print anagrams together in Python using List and Dictionary

```python
words = ["eat", "tea", "tan", "ate", "nat", "bat"]

anagrams = {}

for word in words:
    # Sort the string to create a unique key for all anagrams
    # e.g., 'eat', 'tea', and 'ate' all become 'aet'
    sorted_key = "".join(sorted(word))
    
    if sorted_key not in anagrams:
        anagrams[sorted_key] = []
    
    # Add the original word to the list for that key
    anagrams[sorted_key].append(word)

# Print the grouped lists
print(list(anagrams.values()))
```
